# Importação das Bibliotecas Necessárias

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Carregamento da Base de Dados

```python
df = pd.read_csv("Dados.YT.csv")
df.head()
```

# Estrutura Geral

```python
df.info()
```

# Estatísticas Descritivas

```python
df.describe()
```

# Listagem das Colunas

```python
df.columns
```

# Análise de Valores Ausentes

```python
df.isnull().sum()
```

# Padronização dos Nomes das Colunas

```python
df = df.rename(columns={
    'view_count': 'Views',
    'channel_follower_count': 'Followers',
    'duration': 'Duration',
    'duration_string': 'DurationString',
    'live_status': 'LiveStatus',
    'categories': 'Category',
    'tags': 'Tags'
})
df.columns
```

# Remoção de Registros Duplicados

```python
df = df.drop_duplicates()
df.shape
```

# Reavaliação de Valores Ausentes Após Limpeza

```python
df.isnull().sum()
```

# Tratamento de Valores Ausentes na Coluna Views

```python
df = df.dropna(subset=['Views'])
```

# Criação da Métrica de Engajamento

```python
df['Engagement'] = df['Views'] / (df['Followers'] + 1)
df[['Views','Followers','Engagement']].head()
```

# Definição da Variável IsPopula

```python
median_views = df['Views'].median()
df['IsPopular'] = df['Views'] > median_views
df[['Views','IsPopular']].head()
```

# Normalização de Textos (Título e Categoria)

```python
df['Title'] = df['title'].astype(str).str.lower().str.strip()
df['Category'] = df['Category'].astype(str).str.lower().str.strip()
```

# Frequência de Categorias

```python
df['Category'].value_counts()
```

# Top 10 Vídeos com Mais Visualizações

```python
df.nlargest(10, 'Views')[['Title', 'Views', 'Category']]
```

# Média de Visualizações por Categoria

```python
df.groupby('Category')['Views'].mean().sort_values(ascending=False)
```

# Engajamento Médio por Categoria

```python
df.groupby('Category')['Engagement'].mean().sort_values(ascending=False)
```

# Distribuição de Vídeos Populares

```python
df['IsPopular'].value_counts()
```

# Categorias com Maior Número de Vídeos Populares

```python
df[df['IsPopular'] == True]['Category'].value_counts()
```

# Distribuição das Visualizações (Histograma)

Este gráfico mostra como os valores de visualizações estão distribuídos no dataset.  
Ele ajuda a identificar se há concentração de vídeos com poucas views ou muitos outliers (vídeos extremamente populares).

```python
plt.figure(figsize=(10,5))
sns.histplot(df['Views'], bins=50, kde=True)
plt.title("Distribuição de Visualizações")
plt.xlabel("Views")
plt.ylabel("Frequência")
plt.show()
```

# Média de Visualizações por Categoria

Aqui analisamos quais categorias apresentam maior média de visualizações, permitindo identificar temas que tendem a gerar mais interesse do público.

```python
plt.figure(figsize=(12,6))
df.groupby('Category')['Views'].mean().sort_values(ascending=False).plot(kind='bar')
plt.title("Média de Views por Categoria")
plt.ylabel("Média de Views")
plt.xlabel("Categoria")
plt.xticks(rotation=45)
plt.show()
```

# Engajamento Médio por Categoria

Comparamos o engajamento médio entre as categorias, permitindo entender quais nichos entregam melhor desempenho relativo ao tamanho do canal.

```python
plt.figure(figsize=(12,6))
df.groupby('Category')['Engagement'].mean().sort_values(ascending=False).plot(kind='bar', color='orange')
plt.title("Engajamento Médio por Categoria")
plt.ylabel("Engajamento")
plt.xlabel("Categoria")
plt.xticks(rotation=45)
plt.show()
```

# Distribuição de Views por Categoria (Boxplot)

O boxplot permite visualizar a variação e a presença de outliers nas visualizações dentro de cada categoria.

```python
plt.figure(figsize=(14,7))
sns.boxplot(x='Category', y='Views', data=df)
plt.title("Distribuição de Views por Categoria")
plt.xticks(rotation=45)
plt.show()
```

# Comparação entre Vídeos Populares e Não Populares

Com base na variável `IsPopular`, comparamos a quantidade de vídeos acima e abaixo da mediana de visualizações.

```python
plt.figure(figsize=(6,4))
df['IsPopular'].value_counts().plot(kind='bar', color=['green','red'])
plt.xticks([0,1], ['Não Popular','Popular'])
plt.title("Quantidade de Vídeos Populares vs Não Populares")
plt.ylabel("Quantidade")
plt.show()
```

# Conclusões Finais

Após a análise exploratória, pré-processamento, consultas e visualizações realizadas no dataset de vídeos do YouTube, foi possível identificar insights importantes sobre o comportamento das publicações e seu desempenho.

## 1. Distribuição das Visualizações
A distribuição de `Views` mostrou forte assimetria à direita, evidenciando que:
- a maior parte dos vídeos recebe poucas visualizações;
- poucos vídeos concentram visualizações extremamente altas (outliers);
- isso é típico de plataformas de conteúdo, onde poucos vídeos viralizam.

## 2. Desempenho por Categoria
A análise de categorias revelou que algumas categorias apresentam médias de visualizações significativamente maiores do que outras, indicando que:
- certos nichos tendem a atrair mais público;
- é possível identificar temas mais “quentes” na plataforma;
- categorias pouco populares apresentam menor variação e baixa média de views.

## 3. Engajamento Relativo ao Tamanho do Canal
A métrica de engajamento criada (**Views / Followers**) mostrou que:
- canais pequenos podem ter engajamento proporcional maior;
- alguns vídeos performam muito acima do esperado para o número de seguidores;
- a métrica permite comparar canais de tamanhos diferentes de forma justa.

## 4. Popularidade dos Vídeos
A variável `IsPopular`, baseada na mediana de `Views`, mostrou que:
- aproximadamente metade dos vídeos fica abaixo da mediana;
- as categorias com mais vídeos populares também tendem a apresentar maiores médias de visualização;
- isso reforça o papel do nicho na performance do vídeo.

## 5. Qualidade dos Dados e Limitações
Durante o pré-processamento, notou-se que:
- o dataset não possui dados de datas, impedindo análises temporais;
- não há informações de likes/comentários, limitando métricas de engajamento tradicionais;
- apesar disso, foi possível extrair boas informações a partir das variáveis disponíveis.

## Conclusão Geral
A análise permitiu compreender como diferentes categorias, tamanhos de canal e características do conteúdo influenciam os resultados de visualizações.  
Mesmo com limitações, o dataset fornece insights relevantes sobre padrões de consumo e desempenho de vídeos no YouTube.  
Os resultados obtidos servem como base para estudos futuros envolvendo tendências, engajamento e estratégia de conteúdo.
